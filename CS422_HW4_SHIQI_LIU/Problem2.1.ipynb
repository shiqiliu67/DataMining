```python
#All necessary imports
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.impute import SimpleImputer
#from sklearn.datasets.samples_generator import make_blobs
#from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
#Load the auto-mpg sample dataset
#autompg_ds = pd.read_csv('auto-mpg.data', na_values=["?"])
#autompg_ds.head()
autompg_ds =  pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data", delim_whitespace = True, names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin'], na_values=['?'])
```

```python
autompg_ds.head()
```

```python
autompg_ds = autompg_ds.drop(columns=['cylinders', 'model year', 'origin'])
```

```python
autompg_ds.describe()
```

```python
#Impute any missing values with the mean of the dataset
#autompg_ds = autompg_ds,replace('?', np.NaN)
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_mean.fit(autompg_ds)
autompg_ds[autompg_ds.columns] = imp_mean.fit_transform(autompg_ds)
```

```python
autompg_ds.describe()
```

```python
clustering = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='average').fit(autompg_ds)
```

```python
labels = clustering.labels_
print(labels)
```

```python
autompg_ds['cluster'] = clustering.labels_
autompg_ds.describe()
```

```python
#description for the first cluster
c0 = autompg_ds.loc[autompg_ds['cluster'] == 0]
c0.describe()
```

```python
#description for the second cluster
c1 = autompg_ds.loc[autompg_ds['cluster'] == 1]
c1.describe()
```

```python
#description for the third cluster
c2 = autompg_ds.loc[autompg_ds['cluster'] == 2]
c2.describe()
```
