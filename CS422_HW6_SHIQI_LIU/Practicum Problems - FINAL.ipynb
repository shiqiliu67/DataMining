```python
new_dict = dict()
cosine_similarity_list = []
similar_users_list = []
```

```python
user_cols = ['user_id', 'age', 'gender', 'occupation', 'zip_code']
user_data = pd.read_csv('ml-100k/u.user', 
                        sep='|', names = user_cols)
```

```python
rating_cols = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings_data = pd.read_csv('ml-100k/u.data', 
                           sep='\t', names = rating_cols)
```

```python
item_cols = ['movie id', 'movie title', 'release date', 'video release date', 'IMDb URL', 'Unknown', 'Action', 'Adventure', 'Animation',
             'Childrens', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'FilmNoir', 'Horror', 'Musical', 'Mystery', 'Romance',
             'SciFi', 'Thriller', 'War', 'Western']

items_data = pd.read_csv('ml-100k/u.item', 
                         sep='|', names=item_cols, encoding = 'latin-1')
```

```python
utility = ratings_data.pivot(index='user_id', columns='movie_id', values='rating')
user_means = utility.mean(axis=1)
```

```python
utility_centered = utility - user_means
utility_centered = utility_centered.where((pd.notnull(utility_centered)), 0)
```

```python
print(utility_centered)
```

```python
user_1 = utility_centered[0:1]
user_1.head()
```

```python
for index, row in utility_centered.iterrows():
    np_row = np.array(row)
    np_row.resize(1, 1682)
    if index != 1:
        list = metrics.pairwise.cosine_similarity(user_1, np_row)
        cosine_similarity_list.append(list)
        new_dict.__setitem__(index, list)
```

```python
import heapq
similar_users_top_ten = heapq.nlargest(10, new_dict, key=new_dict.get)
user_columns = ['user_id' , '508']
```
